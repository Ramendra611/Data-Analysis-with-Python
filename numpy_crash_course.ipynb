```python
import time
# Python list — slow loop-based
python_list = list(range(1_000_000_00))
start = time.time()
result = [x * 2 for x in python_list]
print(f"List: {time.time() - start:.4f}s")

# NumPy array — vectorized C operations
np_array = np.arange(1_000_000_00)
start = time.time()
result = np_array * 2
print(f"NumPy: {time.time() - start:.4f}s")
```

# np.where --> important concept

```python
nums = np.array(([54, 62, 79, 105, 40, 201 ]))


# we are proving only the condition
np.where( nums %2 == 0 )  # one way of using the np.where
```

```python
# we are providing the condition and the values to take when condition is true and when conditon is false 


'''

condition  is nums %2 == 0

where condition is true it will take the value from 100
where condition is false it will take the values from nums

'''

np.where( nums %2 == 0, 100, nums)
```

```python
nums = np.array(([54, 62, 79, 105, 40, 201 ]))
backup = np.array(([-100, -362, -379, -405, -340, -901 ]))
backup2 = backup*10000

'''
where the condition is true take the value from backup
where the condition is false take the value from nums

'''
np.where( nums %2 == 0, backup, nums)
```

```python
nums = np.array  (([ 54,  62,   79,   105,   40,  201 ]))
backup = np.array(([-100, -362, -379, -405, -340, -901 ]))
backup2 = backup*10000

'''
where the condition is true take the value from backup
where the condition is false take the value from backup

'''
np.where( nums %2 == 0, backup, backup2)
```
