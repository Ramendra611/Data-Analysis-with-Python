```python
my_table = pd.DataFrame( {
    "name" : ['rohan', 'mohan', 'shyam', 'ghanshyam', "sohan"], 
    "city" : ['delhi', 'mumbai', 'delhi', 'hyderabad', "mumbai"], 
    "salary": [10000, 12000, 15000, 9000 , 18000], 
    "deptt": ["finance", "marketing", "engineering", "HR", "admin"]
    
}   )


bonus_table = pd.DataFrame({ "area" : [ "delhi", "mumbai" , "kolkata"], 
                           "bonus_amount": [0.2, 0.1, 0.3]})

print(bonus_table)

print(my_table)
```

```python
my_table
```

```python
## only those rows which are present in both left and right will be kept
pd.merge( left = my_table, right= bonus_table, how = "inner", left_on = "city", right_on="area")
```

```python
## left join ---> it will keep all the rows from the left table 
pd.merge( left = my_table, right= bonus_table, how = "left", left_on = "city", right_on="area")
```

```python
## right join ---> it will keep all the rows from the right table 
pd.merge( left = my_table, right= bonus_table, how = "right", left_on = "city", right_on="area")
```

```python
## outer join ---> it will keep all the rows from both the sides 
pd.merge( left = my_table, right= bonus_table, how = "outer", left_on = "city", right_on="area")
```

```python
merged  = pd.merge( left = my_table, right= bonus_table, how = "left", left_on = "city", right_on="area")
```

```python
merged
```

```python
## handle the missing value here by imputing a default bonus of 5%

merged["bonus_amount"] = merged["bonus_amount"].apply( lambda x: x if pd.notnull(x) else 0.05 )
```

```python
merged
```

```python
merged["bonus_amount"]
```

```python
merged["updated_bonus"] =   np.where(pd.notnull(merged["bonus_amount"]), merged["bonus_amount"], 0.05 ) # much faster than using apply
```

```python
merged
```

```python
merged["updated_bonus2"]  =  merged["bonus_amount"].fillna(0.02)
merged
```

```python
merged["new_salary"] =  merged["salary"] * (1+merged["updated_bonus2"])
merged
```

```python
## till now we have used apply on a single column. We can also use apply on multiple columns

merged[["salary", "updated_bonus2"]].apply(lambda x : x[0] * (1+x[1]) ,axis = 1)
```

```python
merged["appraised_salary"] =   merged.apply(lambda x : x["salary"] * (1+x["updated_bonus2"]) ,axis = 1)
merged
```

## groupby and aggregation

```python
emp_table
```

```python
## how many employees are there in each city. 

emp_table.groupby(["city"])[["name"]].count()


```

```python
## how many employees working in each department
emp_table.groupby("deptt")[["name"]].count()
```

```python
## average salary city wise
emp_table.groupby("city")["salary"].mean()
```

```python
# filter the data for delhi and find the average
emp_table[emp_table["city"] == "mumbai"]["salary"].mean()
```

```python
## what is the min and max salary in each city

pd.concat([emp_table.groupby("city")["salary"].max(), 
           emp_table.groupby("city")["salary"].min()], axis = 1).reset_index()
```

```python
emp_table.groupby("city")["salary"].agg(["min", "max", "mean"]).reset_index()
```

```python
emp_table
```

```python
## How many employes earn more than average salary


# this will filter the data for those who earn more than the average
emp_table[emp_table["salary"] > np.mean(emp_table["salary"])]

more_than_average = (emp_table[emp_table["salary"] > emp_table["salary"].mean()])


## i want to create flag that is True if person earns more than the average else False
emp_table["salary_more_than_average"] =   emp_table["salary"].apply(lambda x : True if x > np.mean(emp_table["salary"]) else False)
emp_table
```

```python
## Add a new column that shows the average salary of that city

emp_table
```

```python
city_average = emp_table.groupby("city")[["salary"]].mean().reset_index()
city_average = city_average.rename({"salary":"avg_city_salary"}, axis = 1)
city_average
```

```python
pd.merge(emp_table, city_average, left_on= "city", right_on="city", how = "left")
```

```python
merged = emp_table.join(city_average, how = "left",  lsuffix="x", rsuffix="y")
merged
```

```python
## how many employees are earning more than the average salary of their city ( groupby and join or groupby transform )


merged["earns_more_than_city_avg"] =   merged["salary"] > merged["avg_city_salary"]
merged
```
